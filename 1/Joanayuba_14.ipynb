```python
import numpy as np
```

```python
visits = np .array ([120,135,150,155,170,180,175,190,200,210,205,220,230])
print(visits)
 #Type of the dataset
type_of_data = visits.dtype

# Total number of elements
num_elements = visits.size

# Total number of visits
total_visits = visits.sum
```

```python
 # TASK 2 
# Average daily visits
average_visits = visits.mean()
print(average_visits)
# Highest visits
highest_visits = visits.max()
print(highest_visits)
# Lowest visits
lowest_visits = visits.min()
print(lowest_visits)
# Standard deviation
std_deviation = visits.std()
print(std_deviation)
```

```python
# TASK 3 
# increase visit by 15%
increased_visits = visits * 1.15
print(increased_visits)
# min-max normalization
normalized_visits = (increased_visits - increased_visits.min()) / \
                    (increased_visits.max() - increased_visits.min())
print(normalized_visits )
# round to 2 decmical places
rounded_normalized = np.round(normalized_visits, 2)
print(rounded_normalized)
```

```python
# TASK 4 
# days above average
above_average_days = visits[visits > average_visits]
print(above_average_days)
# count of days above average
count_above_average = above_average_days.size 
print(count_above_average)
# percentage of days above average
percentage_above_average = (count_above_average / visits.size) * 100
print(percentage_above_average)
```
