### Task 2: 
- The average daily visits
- The highest number of visits
- The lowest number of visits
- The standard deviation of visits


```python
Average = np.average(visits)
```

```python
print('Average daily visits: ', Average)
```

#### Another Method for Average

```python
average = np.round(visits.mean(), 1)
highest_num = visits.max()
lowest_num = visits.min()
STD = np.round(visits.std(), 2)
```

```python
print('VISITS ANALYSIS REPORT')
print('......................................')


print('Average is: ', average)
print('Higest number of visits: ', highest_num)
print('Lowest number of visits: ', lowest_num)
print('Standard Deviation is: ', STD)
```

### Task 3: 
- Increase all daily visits by 15% to simulate marketing growth

```python
Increased_Daily_Visits = visits*1.15
```

```python
print("Increased daily visits: ", Increased_Daily_Visits)
```

- Normalize the data using Min-Max scaling

```python
min_value = Increased_Daily_Visits.min()
max_value = Increased_Daily_Visits.max()

normalized_visits = (Increased_Daily_Visits - min_value / (max_value - min_value))
```

```python
print(normalized_visits)
```

- Round the normalized values to 2 decimal places

```python
Normalized_value = np.round(normalized_visits, 2)
```

```python
Normalized_value
```

### Task 4: 
- Find all days where visits were above the average



```python
# Visits Above Average (VAA)

VAA =  visits[visits > average]
```

```python
print('Visits above average: ', VAA)
```

- Count how many days had visits above average

```python
# Days having Visits Above Average (DVAA)
DVAA = VAA.size
```

```python
print('Days having visits above average: ', DVAA)
```

- Calculate the percentage of days with above-average visits

```python
# Percentage of days with visits abov average
```

```python
percentage = (DVAA*100)/visits.size
```
