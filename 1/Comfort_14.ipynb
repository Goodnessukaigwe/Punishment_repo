### ***Task 1:***

```python
import numpy as np
```

```python
visits = np.array([120, 135, 150, 160, 155, 170, 180, 175, 190, 200, 210, 205, 220, 230])
print(f"Type of dataset: {type(visits)}")
print(f"Total number of elements: {visits.size}")
print(f"Total number of visits: {np.sum(visits)}")
```

### ***Task 2:***

```python
average_visits = np.mean(visits)
maximum_visits = np.max(visits)
minimum_visits = np.min(visits)
standard_deviation = np.std(visits)

print(f"The average daily visits: {average_visits:.2f}")
print(f"The highest number of visits: {maximum_visits}")
print(f"The lowest number of visits: {minimum_visits}")
print(f"The standard deviation of visits: {standard_deviation:.2f}")
```

### ***Task 3:***

```python
Increased_visits = visits * 1.15
print(f"all daily visits by 15%: {Increased_visits}\n")

normalized_visits = (visits - visits.min()) / (visits.max() - visits.min())
print(f"Normalize data using Min-Max scaling: {normalized_visits}\n")

rounded_values = np.round(normalized_visits, 2)
print(f"Normalized values to 2 decimal places: {rounded_values}")
```

### ***Task 4:***

```python
above_average = visits > average_visits
print(f"All days where visits were above the average: {average_visits:.2f}\n")

count = np.sum(above_average)
print(f"Total number of days had visits above average: {count}\n")

percentage = (count / visits.size) * 100
print(f"Percentage of days with above-average visits: {percentage}")
```
